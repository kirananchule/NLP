```python
message.rename(columns={'v1': 'label', 'v2': 'message'}, inplace=True)
```

```python
message.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
message
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(message)):
    review = re.sub(r'[^a-zA-Z0-9\s]','',message['message'][i]) ##applied regular expression
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
    
    
```

```python
corpus
```

```python
##create is bag of word to convert into vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 100,binary=True)  ##max_feature = slection top of vocabulary, binary = true if any number other than 0 and 2 it will consider it as 1
```

```python
x = cv.fit_transform(corpus)
```

```python
# Convert the BoW matrix to a dense array to view the result
bag_of_words = x.toarray()
bag_of_words
```

```python
bag_of_words.shape
```

```python
# Get the feature names (words in the vocabulary)
vocabulary = cv.get_feature_names_out()
vocabulary
```

```python
cv.vocabulary_
```

## BOW using N-gram

```python
##bag of word using NGRAM 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 100,binary=True,ngram_range=(2,3))
```

```python
x = cv.fit_transform(corpus)
```

```python
bag_of_words = x.toarray()
```

```python
vocabulary = cv.get_feature_names_out()
vocabulary
```
